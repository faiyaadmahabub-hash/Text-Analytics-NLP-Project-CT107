Q2. Form Word Stemming

```python
# Loading Data_1.txt

with open("../Data/Data_1.txt", "r", encoding="utf-8") as file:
    text = file.read()

print(text)
```

Tokenization

```python
# import re to to search for pattern in word
import re

# import nltk to work with language & text data
import nltk

# import word_tokenize which splits the text into words & punctuation
from nltk.tokenize import word_tokenize

# download punkt dataset - required to split the text into words as it containes rule for punctuation & sentence boundaries
nltk.download('punkt')
nltk.download('punkt_tab')

# tokenize the text
tokens = word_tokenize(text)

# print tokens (checking purposes)
print(tokens)
```

Demonstrate Word Stemming Using Regular Expression

```python
# define the regular expression stemmer function
def regex_stem(word):
    pattern = r'^(.*?)(ing|ly|ed|ious|ies|ive|es|s|ment)?$'
    stem, suffix = re.findall(pattern, word)[0]
    return stem

# print stemmer results
print([regex_stem(word) for word in tokens])
```

Demonstrate Word Stemming Using Porter Stemmer

```python
# imports port stemmer from nltk - it's a stemming algorithm the removes common endings
from nltk.stem import PorterStemmer

# initialize porter stemmer - creating a port stemmer object
ps = PorterStemmer()

# this will stem each word and print the result
print([ps.stem(word) for word in tokens])
```

Demonstrate Word Stemming Using Lancaster Stemmer

```python
# imports lacaster stemmer from nltk
from nltk.stem import LancasterStemmer

# initialize lancaster stemmer - creating a lancaster stemmer object
ls = LancasterStemmer()

# this will stem each word and print the result
print([ls.stem(word) for word in tokens])
```
